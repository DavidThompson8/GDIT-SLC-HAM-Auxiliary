```python
import os
import shutil
import csv
import random
```

```python
#path variables
cwd = os.getcwd()
ham = cwd + '\\HAM10000data'
train = cwd + '\\HAM10000data\\train'
val = cwd + '\\HAM10000data\\validation'

#class path modifiers
nv = '\\nv\\'
mel = '\\mel\\'
bkl = '\\bkl\\'
bcc = '\\bcc\\'
akiec = '\\akiec\\'
vasc = '\\vasc\\'
df = '\\df\\'

#file extension
jpg = '.jpg'

#number of images per class and number to be used in training
nv_train = 5364    #5364
nv_total = 6705      #1341
mel_train = 890    #890
mel_total = 1113      #223
bkl_train = 879    #879
bkl_total = 1099      #220
bcc_train = 411    #411
bcc_total = 514      #103
akiec_train = 262  #262
akiec_total = 327     #65
vasc_train = 113   #113
vasc_total = 142     #29
df_train = 92      #92
df_total = 115        #23

#positions in arrays
nv_count = 0
mel_count = 0
bkl_count = 0
bcc_count = 0
akiec_count = 0
vasc_count = 0
df_count = 0
```

```python
#build arrays of values
nv_list = []
for i in range(nv_total):
    nv_list.append(i)
    random.shuffle(nv_list)

mel_list = []
for i in range(mel_total):
    mel_list.append(i)
    random.shuffle(mel_list)

bkl_list = []
for i in range(bkl_total):
    bkl_list.append(i)
    random.shuffle(bkl_list)

bcc_list = []
for i in range(bcc_total):
    bcc_list.append(i)
    random.shuffle(bcc_list)

akiec_list = []
for i in range(akiec_total):
    akiec_list.append(i)
    random.shuffle(akiec_list)

vasc_list = []
for i in range(vasc_total):
    vasc_list.append(i)
    random.shuffle(vasc_list)


df_list = []
for i in range(df_total):
    df_list.append(i)
    random.shuffle(df_list)
```

```python
def move(type, id):
    #class image counts
    global nv_train, nv_total, mel_train, mel_total, bkl_train, bkl_total,\
    bcc_train, bcc_total, akiec_train, akiec_total, vasc_train, vasc_total, df_train, df_total, \
    nv_count, mel_count, bkl_count, bcc_count, akiec_count, vasc_count, df_count
    
    #address of current image
    curr = ham + '\\' + id + jpg
    
    #only attempt to sort if the image exists
    if os.path.exists(curr):
    
    #sort image based on class type
    #list values lower than the target are selected for training, others are used for validation
        if type == 'nv':
            if nv_train > nv_list[nv_count]:
                dest = train + nv + id + jpg
                shutil.move(curr, dest)
            else:
                dest = val + nv + id + jpg
                shutil.move(curr, dest)
            nv_count += 1
        elif type == 'mel':
            if mel_train > mel_list[mel_count]:
                dest = train + mel + id + jpg
                shutil.move(curr, dest)
            else:
                dest = val + mel + id + jpg
                shutil.move(curr, dest)
            mel_count += 1
        elif type == 'bkl':
            if bkl_train > bkl_list[bkl_count]:
                dest = train + bkl + id + jpg
                shutil.move(curr, dest)
            else:
                dest = val + bkl + id + jpg
                shutil.move(curr, dest)
            bkl_count += 1
        elif type == 'bcc':
            if bcc_train > bcc_list[bcc_count]:
                dest = train + bcc + id + jpg
                shutil.move(curr, dest)
            else:
                dest = val + bcc + id + jpg
                shutil.move(curr, dest)
            bcc_count += 1
        elif type == 'akiec':
            if akiec_train > akiec_list[akiec_count]:
                dest = train + akiec + id + jpg
                shutil.move(curr, dest)
            else:
                dest = val + akiec + id + jpg
                shutil.move(curr, dest)
            akiec_count += 1
        elif type == 'vasc':
            if vasc_train > vasc_list[vasc_count]:
                dest = train + vasc + id + jpg
                shutil.move(curr, dest)
            else:
                dest = val + vasc + id + jpg
                shutil.move(curr, dest)
            vasc_count += 1
        elif type == 'df':
            if df_train > df_list[df_count]:
                dest = train + df + id + jpg
                shutil.move(curr, dest)
            else:
                dest = val + df + id + jpg
                shutil.move(curr, dest)
            df_count += 1
```

```python
#use metadata file to extract image classes and sort accordingly
with open('HAM10000_metadata.csv', newline='') as csv_file:
    csv_reader = csv.reader(csv_file, delimiter=',')
    first_line = True
    for row in csv_reader:
        type = row[2]
        id = row[1]
        move(type, id)
```

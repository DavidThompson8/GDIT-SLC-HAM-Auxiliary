```python
import os
import shutil
import csv
```

```python
#path variables
cwd = os.getcwd()
ham = cwd + '\\HAM10000data'
train = cwd + '\\HAM10000data\\train'
val = cwd + '\\HAM10000data\\validation'

#class path modifiers
nv = '\\nv\\'
mel = '\\mel\\'
bkl = '\\bkl\\'
bcc = '\\bcc\\'
akiec = '\\akiec\\'
vasc = '\\vasc\\'
df = '\\df\\'

#file extension
jpg = '.jpg'

#limits for number of images to sort
#can be changed to use a smaller/larger sample size if desired (example use- faster model training for demo purposes)
#commented values are the 80/20 distribution used in the actual SLC model
nv_train = 5364    #5364
nv_val = 1341      #1341
mel_train = 890    #890
mel_val = 223      #223
bkl_train = 879    #879
bkl_val = 220      #220
bcc_train = 411    #411
bcc_val = 103      #103
akiec_train = 262  #262
akiec_val = 65     #65
vasc_train = 113   #113
vasc_val = 29     #29
df_train = 92      #92
df_val = 23        #23
```

```python
def move(type, id):
    #class image counts
    global nv_train, nv_val, mel_train, mel_val, bkl_train, bkl_val,\
    bcc_train, bcc_val, akiec_train, akiec_val, vasc_train, vasc_val, df_train, df_val
    
    #address of current image
    curr = ham + '\\' + id + jpg
    
    #only attempt to sort if the image exists
    if os.path.exists(curr):
    
    #sort image based on class type
    #when training quota is met, add to validation
    #if validation quota is met, do not sort image
        if type == 'nv':
            if nv_train > 0:
                dest = train + nv + id + jpg
                shutil.move(curr, dest)
                nv_train -= 1
            elif nv_val > 0:
                dest = val + nv + id + jpg
                shutil.move(curr, dest)
                nv_val -= 1
        elif type == 'mel':
            if mel_train > 0:
                dest = train + mel + id + jpg
                shutil.move(curr, dest)
                mel_train -= 1
            elif mel_val > 0:
                dest = val + mel + id + jpg
                shutil.move(curr, dest)
                mel_val -= 1
        elif type == 'bkl':
            if bkl_train > 0:
                dest = train + bkl + id + jpg
                shutil.move(curr, dest)
                bkl_train -= 1
            elif bkl_val > 0:
                dest = val + bkl + id + jpg
                shutil.move(curr, dest)
                bkl_val -= 1
        elif type == 'bcc':
            if bcc_train > 0:
                dest = train + bcc + id + jpg
                shutil.move(curr, dest)
                bcc_train -= 1
            elif bcc_val > 0:
                dest = val + bcc + id + jpg
                shutil.move(curr, dest)
                bcc_val -= 1
        elif type == 'akiec':
            if akiec_train > 0:
                dest = train + akiec + id + jpg
                shutil.move(curr, dest)
                akiec_train -= 1
            elif akiec_val > 0:
                dest = val + akiec + id + jpg
                shutil.move(curr, dest)
                akiec_val -= 1
        elif type == 'vasc':
            if vasc_train > 0:
                dest = train + vasc + id + jpg
                shutil.move(curr, dest)
                vasc_train -= 1
            elif vasc_val > 0:
                dest = val + vasc + id + jpg
                shutil.move(curr, dest)
                vasc_val -= 1
        elif type == 'df':
            if df_train > 0:
                dest = train + df + id + jpg
                shutil.move(curr, dest)
                df_train -= 1
            elif df_val > 0:
                dest = val + df + id + jpg
                shutil.move(curr, dest)
                df_val -= 1
```

```python
#use metadata file to extract image classes and sort accordingly
with open('HAM10000_metadata.csv', newline='') as csv_file:
    csv_reader = csv.reader(csv_file, delimiter=',')
    first_line = True
    for row in csv_reader:
        type = row[2]
        id = row[1]
        move(type, id)
```
